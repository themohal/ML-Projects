# Preprocessing

```python
movies.iloc[0].genres
```

```python
import json
# Converting string to list
res = json.loads(movies.iloc[0].genres)
res[0]['name']
```

```python
def convert(obj):
  res = json.loads(obj)
  L = []
  for i in range(len(res)):
    L.append(res[i]['name'])
  return L
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.cast[0]
```

```python
def convert_cast(obj):
  res = json.loads(obj)
  count = 0
  L = []
  for i in range(len(res)):
      if count !=3:
        L.append(res[i]['name'])
        count+=1
      else:
        break
  return L
```

```python
convert_cast(movies['cast'][0])
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
def fetch_director(obj):
  L = []
  res = json.loads(obj)
  for j in range(len(res)):
    if res[j]['job'] =='Director':
      L.append(res[j]['name'])
  return L


```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

## removing spaces
## e.g Sam Worthington to SamWorthington

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

## Concatenate All

```python
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

## Stemming e.g [loved,loving,love]
## will become [love, love,love]

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

## Calculate distance between movies or the vector of movies 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```
